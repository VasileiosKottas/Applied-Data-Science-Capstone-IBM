```python
from folium.plugins import MarkerCluster

# Sample data for rocket launches (replace with actual data)
launch_records = spacex_df.to_dict(orient='records')


# Create a map centered at a specific location
site_map = folium.Map(location=[37.7749, -122.4194], zoom_start=5)

# Create a MarkerCluster object
marker_cluster = MarkerCluster()

# Add markers for each launch record
for record in launch_records:
    # Determine marker color based on launch success
    marker_color = 'green' if record["class"] == 1 else 'red'
    
    # Create a marker with custom icon color
    marker = folium.Marker(
        location=[record["Lat"], record["Long"]],
        icon=folium.Icon(color='white', icon_color=marker_color)
    )
    
    # Add marker to the cluster
    marker_cluster.add_child(marker)

# Add the marker cluster to the map
site_map.add_child(marker_cluster)
```

## TASK 3: Calculate the distances between a launch site to its proximities


```python
# Add Mouse Position to get the coordinate (Lat, Long) for a mouse over on the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# Coordinates of the closest coastline point (replace with your coordinates)
coastline_lat = 34.63283
coastline_lon = -120.6108
launch_site_lat = 34.63166
launch_site_lon = -120.6107
rail_lat = 34.63207
rail_lon = -120.62531
high_lat = 34.63263
high_lon = -120.62376
city_lat = 34.63885
city_lon = -120.45783
coordinates = [[coastline_lat, coastline_lon], [launch_site_lat, launch_site_lon]]
coordinates_rail = [[rail_lat, rail_lon],[launch_site_lat, launch_site_lon]]
coordinates_high = [[high_lat,high_lon],[launch_site_lat, launch_site_lon]]
coordinates_city = [[city_lat,city_lon ],[launch_site_lat, launch_site_lon]]
# Calculate the distance between the coastline point and the launch site
distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)
distance_rail = calculate_distance(launch_site_lat, launch_site_lon,rail_lat, rail_lon)
distance_high = calculate_distance(launch_site_lat, launch_site_lon,high_lat, high_lon)
distance_city = calculate_distance(launch_site_lat, launch_site_lon,city_lat, city_lon)

# Create a marker for the coastline point
coastline_marker = folium.Marker(
    location=[coastline_lat, coastline_lon],
    icon=folium.DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance_coastline),
    )
)
rail_marker = folium.Marker(
    location=[rail_lat, rail_lon],
    icon=folium.DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance_rail),
    )
)
high_marker = folium.Marker(
    location=[high_lat, high_lon],
    icon=folium.DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance_high),
    )
)
city_marker = folium.Marker(
    location=[city_lat, city_lon],
    icon=folium.DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance_city),
    )
)
lines = folium.PolyLine(locations=coordinates, weight=1)
lines_rail = folium.PolyLine(locations=coordinates_rail, weight=1)
lines_high = folium.PolyLine(locations=coordinates_high, weight=1)
lines_city = folium.PolyLine(locations=coordinates_city, weight=1)
site_map.add_child(lines)
site_map.add_child(lines_rail)
site_map.add_child(lines_high)
site_map.add_child(lines_city)
# Add the coastline marker to the map
site_map.add_child(coastline_marker)
site_map.add_child(rail_marker)
site_map.add_child(high_marker)
site_map.add_child(city_marker)
# Display the map
site_map
```
