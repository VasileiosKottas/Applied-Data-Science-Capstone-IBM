```python
df_sql = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_2.csv')
df_sql.to_sql("SPACEXTBL", con, if_exists='replace', index=False,method="multi")
df_sql.head()
```

```python
# Execute SQL query to calculate the success rate of each orbit
result = %sql SELECT Orbit, AVG(Class) AS Success_Rate FROM SPACEXTBL GROUP BY Orbit

# Convert the result to a DataFrame
df_m = result.DataFrame()

# Plotting the bar chart
plt.figure(figsize=(10, 6))
plt.bar(df_m['Orbit'], df_m['Success_Rate'], color='skyblue')
plt.title('Success Rate of Each Orbit')
plt.xlabel('Orbit')
plt.ylabel('Success Rate')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

## TASK  6: Visualize the launch success yearly trend


```python
# A function to Extract years from the date 
year=[]
def Extract_year():
    for i in df["Date"]:
        year.append(i.split("-")[0])
    return year
Extract_year()
df['Date'] = year
df.head()
```

```python
sns.lineplot(x='Date',y='Outcome',data=df)
plt.xlabel("Date",fontsize=12)
plt.ylabel("Outcome",fontsize=12)
plt.show()
```
